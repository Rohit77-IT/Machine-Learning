```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("House_price.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df['Area (sq ft)'], df['Price ($)'], color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area (sq ft)']], df['Price ($)'])
```

```python
reg.predict([[2300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df['Area (sq ft)'], df['Price ($)'], color='red', marker='+')
plt.plot(df['Area (sq ft)'], reg.predict(df[['Area (sq ft)']]), color='blue')
```
